* 아래에 답을 서술하여 주세요
* `answer4`에 정답을 담아주세요

```python
def sigmoid(logits) :
    return 1 / ( 1 + np.exp(-logits))

def cee(pred, y_true) :
    return -np.mean(y_true*np.log(pred) + (1-y_true)*np.log(1-pred))

X = 3
Y = 1
Loss = 2.6

################################

W0 = 0.12
Z = X * W0 
a = sigmoid(Z)

W1 = 0.07
logits = a * W1
pred = sigmoid(logits)

loss = cee(pred, Y)

################################

dLoss_dy = pred - Y  #cee + sigmoid에 대한 미분(sigmoid의 입력부분까지)
dy_da = W1        # 위에 그림에서 logits은 y와 같은 값임.
da_dz = a * (1-a) # sigmoid에 대한 미분
dz_dw0 = X

###############################

dLoss_dW0 = dLoss_dy * dy_da * da_dz * dz_dw0
answer4 = dLoss_dW0
answer4
```
